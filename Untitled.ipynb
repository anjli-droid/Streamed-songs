```python
import pandas as pd
```

```python
df = pd.read_csv("MostStreamedSpotifySongs2024.csv",encoding="latin1")
```

```python
df.head()
```

```python
#selecting important columns
df = df[[
    "Track",
    "Artist",
    "Album Name",
    "Release Date",
    "ISRC",
    "All Time Rank",
    "Track Score",
    "Spotify Streams",
    "Spotify Playlist Count",
    "Spotify Playlist Reach",
    "Spotify Popularity",
    "YouTube Views",
    "YouTube Likes",
    "Apple Music Playlist Count"
]]
```

```python
#renaming columns
df.columns = ["track_name","artist","album_name","release_date","isrc","all_time_rank","track_score","spotify_streams","spotify_playlist_count","spotify_playlist_reach","spotify_popularity","youtube_views", "youtube_likes","apple_music_playlist_count"]
```

```python
numeric_cols = ["all_time_rank","track_score","spotify_streams","spotify_playlist_count","spotify_playlist_reach","spotify_popularity","youtube_views","youtube_likes", "apple_music_playlist_count"]
```

```python
df[numeric_cols] = (df[numeric_cols].replace(",", "", regex=True).apply(pd.to_numeric, errors="coerce"))
```

```python
df["release_date"] = pd.to_datetime(df["release_date"],errors="coerce")
```

```python
df = df.dropna()
```

```python
df
```

```python
df["spotify_streams"] = df["spotify_streams"].astype(int)
```

```python
df["youtube_views"] = df["youtube_views"].astype(int)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.shape
```

```python
df["song_id"] = df.index + 1
```

```python
songs_dim = df[["song_id","track_name","artist","album_name","release_date","isrc"]].copy()
```

```python
song_metrics = df[["song_id","all_time_rank","track_score","spotify_streams","spotify_playlist_count","spotify_playlist_reach","spotify_popularity","youtube_views","youtube_likes","apple_music_playlist_count"]].copy()
```

```python
songs_dim.to_csv("songs_dim.csv", index=False)
song_metrics.to_csv("song_metrics.csv", index=False)
```
